```python
# Matplotlib Part - 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate a graph using a Dataset
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Grade_Set_1.csv')
data
```

```python
plt.scatter(data.Hours_Studied,data.Test_Grade,color='red')
plt.plot(data.Hours_Studied,data.Test_Grade,color='green',label='Data From Dataset')
plt.xlabel('Hours_Studied',fontsize=12)
plt.ylabel('Marks',fontsize=12)
plt.title('Hours_Studied vs Test_Grade',fontsize=10,fontweight='bold')
plt.legend()
plt.show()
```

```python
# xticks() and yticks()
from numpy import random
x = random.randint(100,size=(10))  # 1D array of 10 elements [random no. b/w 1 till 100]
y = random.randint(60,size=(10))
plt.plot(x,y,color='red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Xticks and Yticks')
plt.xticks(np.arange(0,105,5))
plt.yticks(np.arange(0,60,5))
plt.show()
```

```python
# Histogram
# a histogram is used to show the distribution of data
# for example : if you have the ages of 100 people , a histogram help you see how many people are in their in 20's
# 30s , 40s
# ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
# bins = 5
# divide the range of ages (18-50) into 5 equal sized groups
# 18-24 25-31 32-38 39-45 46-52

ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
plt.hist(ages,bins=5,color='skyblue',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number Of People")
plt.show()
```

```python
# Heatmap
# a heatmap is a colorful table
# instead of showing numbers as text , it will show you colors to make patterns
# easier to spot

# a tables where each cell's color tells you "how big or small" a value is
# darker color for higher value
# lighter color for smaller value

# correlation
# correlation used to find out the relationship

# correlation is measured on a scale of +1 and -1

# imagine you 2 things  a.) Hours Studied  b.) Marks scored
# if a student studies more hours usually more marks  : pos relationship
# if student watches more TV and scores less          : neg relationship

import seaborn as sns

data = {
    'Hours_Studied':[2,3,4,5,6,7,8],
    'Marks_Scored':[40,50,60,70,80,90,100],
    'TV_Watched':[10,9,8,6,7,5,3]
}

df = pd.DataFrame(data)
corr = df.corr()

sns.heatmap(corr,annot=True)
plt.title('HeatMap')
plt.show()
```

```python
x = random.randint(10,size=(7))
y = random.randint(6,size=(7))
color = ['green', 'blue', 'skyblue', 'red', 'orange','yellow']

for i in range(1,7):
  plt.subplot(2,3,i)
  plt.plot(x,y,color=color[i-1])
  plt.title(f'Plot {i}')
  plt.xlabel('X axis')
  plt.ylabel('Y axis')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.7,
                    hspace=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

colors = ['green', 'red', 'blue', 'purple', 'orange', 'brown']

x_max = [10, 20, 15, 12, 18, 25]
y_max = [6, 10, 8, 9, 7, 12]

# Loop --> 6 subplots in 2x3 layout
for i in range(6):
    x = np.random.randint(x_max[i], size=(7))
    y = np.random.randint(y_max[i], size=(7))
    plt.subplot(2, 3, i + 1)  # 2 rows, 3 columns, position i+1
    plt.plot(x, y, color=colors[i])
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title(f'Plot {i + 1}')

# Adjust layout and show all subplots instead of plt.subplots_adjust()
plt.tight_layout()
plt.show()
```

```python
# Introduction To
# The Figure and Axes Concept

# a matplotlib chart is comprises of 2 things
#    a.) fig : the container (whole window or page)
#    b.) ax  : the plot inside it
```

```python
plt.figure(facecolor='yellow')
plt.plot([1,2,3,4],[10,20,25,30])
ax = plt.gca() # get current axes
ax.set_facecolor('lightgrey')
plt.title('Without fig,ax')
plt.show()
```

```python
fig,ax = plt.subplots()
# set the background for figure
fig.patch.set_facecolor('yellow')
ax.plot([1,2,3,4],[10,20,25,30])
ax.set_facecolor('lightgrey')
ax.set_title('With fig,ax')
plt.show()
```
