```python
# Pandas Data Science Library
import numpy as np
import pandas as pd
print(pd.__version__)
```

```python
# numpy array
arr = np.array([10,20,30,40])
print(arr)
print(type(arr))
```

```python
# Pandas Series
# A Series is 1D Labeled array

s = pd.Series([10,20,30,40])
print(type(s))
print(s)
print(s.shape)


# In NumPy --> Only Numbers
# In Series --> Numbers + Labels(index)
```

```python
info = pd.Series(['John' , 28 , 6700 , 'KL'] , index=['name','age','sal','city'])
print(info)
print(info['sal'])
print(info.iloc[2])

# Series is like a Dictionary + array combo
```

```python
# values
print(info.values)
```

```python
# Indexing Slicing and Using Loops
s = pd.Series([10,20,30,40,50,60,70] , index=['val1','val2','val3','val4','val5','val6','val7'])
print(s)
```

```python
# Indexing in Series can be done by using [] or iloc[] or loc[]

print(s.iloc[0])       # for index numbers
print(s.loc['val1'])   # loc used for labels
print(s['val5'])       # [] also used for labels
```

```python
# Slicing   10 20 30 40 50 60 70 
#           0   1  2  3  4  5  6
#          v1  v2  v3 v4 v5 v6 v7
print(s[1:6]) 

print()
# when slicing by labels , the end label is included
print(s['val2':'val6']) 
```

```python
# val1 val4 val7
print(s[['val1','val4','val7']])
```

```python
# conditional selection
print(s[s% 4 == 0])
```

```python
# Using Loops
for i in s:
    print(i)
```

```python
for k,v in s.items():
    print(k,v)
```

```python
for i in enumerate(s):
    print(i)
```

```python
# read_csv()
# used to read a CSV File
df = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
df
```

```python
# finding the sum of salary using Pandas
salaries = df['Salary']
print(type(salaries))
total_sal = salaries.sum()
print(total_sal)
```

```python
# finding the sum of salary using Numpy
salaries = df['Salary'].to_numpy()
print(type(salaries))
total_sal = np.sum(salaries)
print(total_sal)
```
